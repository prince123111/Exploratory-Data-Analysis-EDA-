```python
b=df.groupby("Branch")[["Total",'Unit price','Quantity','Tax 5%','gross margin percentage','cogs']].sum().reset_index()
b
```

```python
c=1
for i in b:
    if i=="Branch":
        continue
    else:
        
        plt.figure(figsize=(14,10))
        plt.subplot(2,3,c)
        #plt.bar(b["Branch"],b[i])
        plt.pie(b[i],labels=b["Branch"],autopct="%f0.2%%")
        plt.legend()
        plt.xlabel("Branches")
        plt.ylabel(i)
        c+=1
plt.show()
```

```python
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
sns.scatterplot(data=b,x="Branch",y="Total")
plt.grid()
plt.subplot(2,2,2)
sns.barplot(data=b,x="Branch",y="Unit price")
plt.grid()
plt.subplot(2,2,3)
sns.barplot(data=b,x="Branch",y="Quantity")
plt.grid()
plt.subplot(2,2,4)
sns.barplot(data=b,x="Branch",y="gross margin percentage")
plt.grid()
plt.suptitle("All Comperision by Branches")
plt.show()
```

```python
plt.figure(figsize=(10,8),facecolor="pink")
plt.subplot(2,2,1)
plt.pie(b["Total"],labels=b["Branch"],autopct="%0.2f%%",radius=1)
plt.grid()
plt.legend()
plt.subplot(2,2,2)
sns.barplot(data=b,x="Branch",y="Unit price")
#plt.pie(b["Unit price"],labels=b["Branch"],autopct="%0.2f%%")
plt.grid()
plt.legend()
plt.subplot(2,2,3)
sns.barplot(data=b,x="Branch",y="Quantity")
#plt.pie(b["Quantity"],labels=b["Branch"],autopct="%0.2f%%")
plt.grid()
plt.legend()
plt.subplot(2,2,4)
plt.pie(b["gross margin percentage"],labels=b["Branch"],autopct="%0.2f%%")
plt.grid()
plt.legend(loc="best")
plt.suptitle("All Comperision by Branches")
plt.show()
```

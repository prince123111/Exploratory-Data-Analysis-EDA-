```python
#plt.figure(figsize=(12,6))
plt.rcParams["figure.figsize"]=(12,6)
sns.barplot(x='Aggregate rating',y="Rating_count",data=rate_count)
plt.title("Counting rate respect to Aggregate Rating",fontweight="bold",fontsize=14)
plt.show()
```

```python
#plt.figure(figsize=(12,6))
plt.rcParams["figure.figsize"]=(12,6)
sns.barplot(x='Aggregate rating',y="Rating_count",data=rate_count,hue='Rating color',palette=['white','red','orange','yellow','green','green'])
plt.title("Counting rate respect to Aggregate Rating",fontweight="bold",fontsize=14)
plt.show()
```

# Observation
